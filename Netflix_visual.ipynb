```python
# Netflix user database - performing number of calculations on this set of data

import pandas as pd
import numpy as np

# Uploading and reading the dataset to check if it's loaded properly
path = '/content/Netflix Userbase.csv'

df_raw = pd.read_csv(path)


df_netflix = df_raw
df_netflix.head(5)
```

```python
import matplotlib.pyplot as plt

distinctive = df_netflix['Subscription Type'].unique()
count = df_netflix['Subscription Type'].value_counts()

print(distinctive)
print(count)

wykres1 = plt.pie(count, labels = distinctive, startangle=140, autopct='%1.1f%%', colors =['olivedrab', 'purple', 'brown'])


start_date = '01-05-2023'
end_date = '31-05-2023'

filtered_data = df_netflix[(df_netflix['Join Date'] >= start_date) & (df_netflix['Join Date'] <= end_date)]
sub_counts = filtered_data['Subscription Type'].value_counts()

wykres2 = sub_counts.plot(kind='bar',color='skyblue')




gender_device = df_netflix.groupby(['Gender', 'Device']).size().unstack()

wykres3 = gender_device.plot(kind = 'bar', stacked=True, figsize=(10, 6))
```
