```python
# Netflix user database - performing number of calculations on this set of data

import pandas as pd
import numpy as np

# Uploading and reading the dataset to check if it's loaded properly
path = '/content/Netflix Userbase.csv'

df_raw = pd.read_csv(path)


df_netflix = df_raw
df_netflix.head(5)
```

```python
# Evaluating the dataset to determine if there are any null values + checking on data types available
df_netflix.info()
df_netflix.describe().T
df_netflix.notnull().values.any()
```

```python
# Let's see where the most users come from with Premium subscription

premium = df_netflix['Subscription Type'] == "Premium"
top_premium_by_country = df_netflix[premium]
top_premium_by_country['Country'].value_counts().nlargest(5)
```

```python
# Checking what is the top 3 youngest nation user wise
average_age = df_netflix.groupby('Country')['Age'].mean().sort_values(ascending=True).nlargest(3)
print(average_age)
```

```python
# What is the most frequently used device among users >50 years old?
df_netflix['Age'] = df_netflix['Age'].astype(int)
elderly = df_netflix[df_netflix['Age'] > 50]

elderly['Device'].value_counts()
```

```python

# Let's distinguish that by gender
elderly.groupby('Gender')['Device'].value_counts()
```

```python
# Let's see what is the central date in the column Join Date and see whether more users joined before or after the middle date
df_netflix["Join Date"] = pd.to_datetime(df_netflix["Join Date"])

maxx = df_netflix["Join Date"].max()
minn = df_netflix["Join Date"].min()

differential = (maxx - minn) / 2
meet_me_halfway = minn + differential
print(differential)
print(meet_me_halfway)

before_mid = df_netflix[df_netflix['Join Date'] < meet_me_halfway]
after_or_on_mid = df_netflix[df_netflix['Join Date'] >= meet_me_halfway].value_counts()

print(len(before_mid))
print(len(after_or_on_mid))
```
