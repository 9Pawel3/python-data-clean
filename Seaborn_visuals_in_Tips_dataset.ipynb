```python
# Let's import seaborn library and one of its dataset to do some visualisation
import seaborn as sns

sns.get_dataset_names()
```

```python
# Loading default dataset of tips

tips = sns.load_dataset("tips")
tips.describe()
tips.info()
```

```python
# Let's see what is the relation between the tip and meal price per day. The meal size constitutes for dot size.

visual_1 = sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size = "size")
```

```python
# Statistical type of plot - histogram for the distribution of the tip with the function line

visual_2 = sns.histplot(tips['tip'], kde=True, bins=15)
visual_2
```

```python
# Distribution plot, which is almost the same as histogram
visual_3 = sns.distplot(tips['tip'], kde=True, bins=15)
```

```python
# Barplot to see average tip by gender
visual_4 = sns.barplot(x="sex", y="tip", data=tips, palette="YlGnBu")
```

```python
# Let's see tips by gender and by day thanks to strip plot
visual_5 = sns.stripplot(x="day", y="tip", data=tips, hue="sex", palette="YlGnBu", dodge=True)
visual_5
```
