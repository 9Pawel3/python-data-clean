```python
# loading essential libraries
import pandas as pd
import numpy as np

# loading dataset of global youtube stats
path = "/content/Global YouTube Statistics.csv"
dff = pd.read_csv(path, encoding='latin1')

# getting rid of a number of columns not impacting the dataset (country, location, employment )
dff.columns
df = dff.iloc[:, :-6]
df

# checking for data types and NaNs
df.info()

# unifyig naming capitalize + replacing spaces with underscore
df.columns = df.columns.str.capitalize()
df.columns = df.columns.str.replace(' ', '_')
df
```

```python
# Let's see columns that amass the most % of nulls
null_counts = df.isnull().sum() / len(df)
round(null_counts,2)

# Deleting columns that >10% data is null
to_be_deleted = ['Subscribers_for_last_30_days', 'Country', 'Abbreviation', 'Country_rank']

df = df.drop(columns=to_be_deleted)
df

```

```python
# Now with minor volume of NaNS, let's deleting rows including null values
df = df.dropna()

df.info()
```

Let's put up a number of calulations on given dataset.

```python
# Calculating average of maximum monthly earning for top 5 and low 5 ranked youtubers
top_5_rank = df[df['Rank'] < 6]
low_5_rank = df.sort_values(by='Rank',ascending=False)

average_top5 = top_5_rank['Highest_yearly_earnings'].mean()
average_low5 = low_5_rank['Highest_yearly_earnings'].mean()

print(round(average_top5,2))
print(round(average_low5,2))
```

```python
# Checking the most top 3 common Category type created in 2018
the_year = df[df['Created_year'] == 2018]
most_popular_2018 = the_year['Category'].value_counts().nlargest(3)

print(most_popular_2018)
```

```python
# Calculate top 10 Youtubers with most videos uploaded this year
when = df[df['Created_year'] == 2023]
most_views_2023 = df.groupby('Youtuber')['Video_views'].sum()

top_10 = most_views_2023.nlargest(10)
print(top_10)

```

```python
# Check the least 5 subscribers per Channel Type
sub_qty = df.groupby('Channel_type')['Subscribers'].sum().nsmallest(5)

print(sub_qty)
```

```python
# See trend for number of subscribers for top 5 channel categories past 3 years
y_2023 = df[df['Created_year'] == 2023]
y_2022 = df[df['Created_year'] == 2022]
y_2021 = df[df['Created_year'] == 2021]

top_5_channel = df.groupby('Channel_type')['Subscribers'].sum().nlargest(5)
top_5_channel

top_5_channel_2023 = y_2023[y_2023['Channel_type'].isin(top_5_channel)]
top_5_channel_2022 = y_2022[y_2022['Channel_type'].isin(top_5_channel)]
top_5_channel_2021 = y_2021[y_2021['Channel_type'].isin(top_5_channel)]

avg_subscribers_2023 = top_5_channel_2023.groupby('Channel_type')['Subscribers'].mean()
avg_subscribers_2022 = top_5_channel_2022.groupby('Channel_type')['Subscribers'].mean()
avg_subscribers_2021 = top_5_channel_2021.groupby('Channel_type')['Subscribers'].mean()


```
